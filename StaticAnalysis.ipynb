# Static analysis

## File Type

```python
result = subprocess.run(["file",sample],capture_output=True,text=True)
print(result.stdout)

with open("filetype.txt" ,mode="w")as f:
    f.write(result.stdout)
```

## Hash

```python
result = subprocess.run(["md5sum",sample],capture_output=True,text=True)
print(result.stdout)

with open("md5.txt" ,mode="w")as f:
    f.write(result.stdout)
```

```python
result = subprocess.run(["sha1sum" , sample],text=True,capture_output=True)
print(result.stdout)

with open("sha1.txt" ,mode="w")as f:
    f.write(result.stdout)
```

```python
result = subprocess.run(["sha256sum" , sample],text=True,capture_output=True)
print(result.stdout)

with open("sha265.txt" ,mode="w")as f:
    f.write(result.stdout)
```
